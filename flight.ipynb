```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('flight_dataset.csv')
df
```

```python
df.columns
```

```python
sns.countplot(data=df, x='Month')
```

```python
plt.scatter(x='Destination', y='Dep_hours', data=df)
plt.scatter(x='Dep_hours', y='Destination', data=df)
plt.xticks(rotation=80)
```

```python
sns.scatterplot(
    data = df,
    x = 'Dep_hours',
    y = 'Price',
    hue = 'Month'
)
```

```python
plt.hist(x='Price', data=df, hatch='//')
plt.show()
```

```python
sns.countplot(data=df, x= 'Airline', hue= 'Month')
plt.xticks(rotation=80)
```

```python
sns.barplot(data=df, x='Airline', y='Price')
plt.xticks(rotation=90)
```

```python
sns.histplot(data= df,
        x= 'Source',
        kde=19,
        bins=5)   
```

```python
sns.histplot(data= df,
        x= 'Airline',
        kde=19,
        bins=5)   
plt.xticks(rotation=90) 
```

```python
top_n =10
df['Price'].value_counts().nlargest(top_n).plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
sns.countplot(data= df, y= 'Destination', hue= 'Total_Stops', fill=24)
```

```python
sns.catplot(
    data= df,
    x= 'Dep_hours',
    y= 'Price',
    hue= 'Month',
)
```

```python
sns.catplot(
    data= df,
    x= 'Month',
    # y= 'Price',
    hue= 'Month',
    kind='count'
)
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(x='Airline',y='Price',data=df.sort_values('Price',ascending=False))
plt.xticks(rotation=60)
```
