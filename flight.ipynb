```python
sns.countplot(data=df, x= 'Airline', hue= 'Month')
plt.xticks(rotation=80)
```

```python
sns.barplot(data=df, x='Destination', y='Price')
plt.xticks(rotation=90)
```

```python
sns.histplot(data= df,
        x= 'Airline',
        kde=19,
        bins=5)   
plt.xticks(rotation=90) 
```

```python
top_n =10
df['Price'].value_counts().nlargest(top_n).plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
sns.catplot(
    data= df,
    x= 'Dep_hours',
    y= 'Price',
    hue= 'Month',
)
```

```python
sns.catplot(
    data= df,
    x= 'Month',
    # y= 'Price',
    hue= 'Month',
    kind='count'
)
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(x='Airline',y='Price',data=df.sort_values('Price',ascending=False))
plt.xticks(rotation=60)
```
